```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,4)
x1 = np.cos((((2*x)-1)/len(x))*np.radians(180))
y1 = np.sin((((2*x)-1)/len(x))*np.radians(180))
```

```python
plt.plot(x1,y1)
```

```python
S_points = [(x1[i],y1[i]) for i in range(len(x1))]
```

```python
vertex=np.arange(1,5)
x2 = np.cos((((2*vertex)-1)/len(vertex))*np.radians(180))
y2 = np.sin((((2*vertex)-1)/len(vertex))*np.radians(180))

T_points = [(x2[i],y2[i]) for i in range(len(x2))]
```

```python
plt.plot(x2,y2)
```

```python
S_points+T_points
```

```python
other_shape_x = [i[0] for i in S_points+T_points]
other_shape_y = [i[1] for i in S_points+T_points]
plt.plot(other_shape_x,other_shape_y)
```
